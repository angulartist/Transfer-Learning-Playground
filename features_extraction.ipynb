### Features extraction + Neural net

```python
from keras.optimizers import RMSprop
from keras import applications

from keras import models, layers

from albumentations import (
    Compose, ShiftScaleRotate
)

from h5imagegenerator import HDF5ImageGenerator
```

```python
# Get the pretrained VGG16 model
# on imagenet and remove the
# fully-connected layers
base = applications.inception_v3.InceptionV3(
    weights='imagenet',
    include_top=False,
    input_shape=(299, 299, 3),
)

# Freeze the convolutional base
base.trainable = False
```

```python
# Extend the base model with
# a new "head" of dense layers
model = models.Sequential()

model.add(base)
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(2, activation='softmax'))
```

```python
num_epochs = 25
batch_size = 32
```

```python
aug = Compose([
    ShiftScaleRotate(
        shift_limit=0.02,
        scale_limit=0.04,
        rotate_limit=15,
    )
])

train_gen = HDF5ImageGenerator(
    src= 'datasets/chest/train.h5',
    scaler=True,
    num_classes=2,
    batch_size=batch_size,
    augmenter=aug,
)

val_gen = HDF5ImageGenerator(
    src= 'datasets/chest/test.h5',
    scaler=True,
    num_classes=2,
    batch_size=batch_size,
)
```

```python
rms = RMSprop(lr=0.001, rho=0.9)

callbacks = []

model.compile(
    loss='categorical_crossentropy',
    metrics=['accuracy'],
    optimizer=rms
)

h = model.fit_generator(
    train_gen,
    validation_data=val_gen,
    workers=10,
    use_multiprocessing=True,
    epochs=num_epochs,
    verbose=1,
)
```
